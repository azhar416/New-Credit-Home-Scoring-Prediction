# Import Data Train dan Test

```python
# load data
X_train = pd.read_csv(root+'X_train_2.csv')
X_test = pd.read_csv(root+'X_test_2.csv')
y_train = pd.read_csv(root+'y_train_2.csv')
y_test = pd.read_csv(root+'y_test_2.csv')
```

# Import Model

```python
# Load Model
LR_Model = pickle.load(open(root + "Model/LR_Model_3.pkl", 'rb'))
RF_Model = pickle.load(open(root + "Model/RF_Model_3.pkl", 'rb'))
KNN_Model = pickle.load(open(root + "Model/KNN_Model_3.pkl", 'rb'))
```

# Train Ensemble

## Hard Voting

```python
# create a voting classifier with hard voting
voting_classifier_hard = VotingClassifier(
    estimators = [('rf', RF_Model),
                  ('lr', LR_Model),
                  ('knn', KNN_Model)],
    voting='hard')
```

```python
# make predictions with the hard voting model
voting_classifier_hard.fit(X_train.values, y_train.values.ravel())
```

```python
# Save Model
pickle.dump(voting_classifier_hard, open(root + "Model/Ensemble_Model_3_Hard.pkl", 'wb'))
```

## Soft Voting

```python
# create a voting classifier with soft voting
voting_classifier_soft = VotingClassifier(
    estimators = [('rf', RF_Model),
                  ('lr', LR_Model),
                  ('knn', KNN_Model)],
    voting='soft')
```

```python
# make predictions with the soft voting model
voting_classifier_soft.fit(X_train.values, y_train.values.ravel())
```

```python
# Save Model
pickle.dump(voting_classifier_soft, open(root + "Model/Ensemble_Model_3_Soft.pkl", 'wb'))
```
