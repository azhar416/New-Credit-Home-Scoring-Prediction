# Import Dataset 

```python
# tabel setting
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
data = pd.read_csv(root+'New_Dataset.csv')
```

```python
data.head()
```

```python
data.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
# Cek Missing Data
total_missing = data.isnull().sum().sort_values(ascending=False)
percent_1 = data.isnull().sum()/data.isnull().count()*100
percent_2 = (round(percent_1, 1)).sort_values(ascending=False)
missing_data = pd.concat([total_missing, percent_2], axis=1, keys=['Total Missing', '%'])
missing_data = missing_data.reset_index().rename(columns={'index': 'Column'})
missing_data
```

# Split Dataset

```python
# Drop Target dan ID
X = data.drop(['SK_ID_CURR','TARGET'], axis=1)
Y = data.TARGET
```

```python
# Split Dataset
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=1, stratify=Y)
```

```python
print('Shape X_train : ', X_train.shape)
print('Shape X_test : ', X_test.shape)
print('Shape y_train : ', y_train.shape)
print('Shape y_test : ', y_test.shape)
```

```python
# save train and test data
X_train.to_csv(root+"X_train.csv", index=False)
X_test.to_csv(root+"X_test.csv", index=False)
y_train.to_csv(root+"y_train.csv", index=False)
y_test.to_csv(root+"y_test.csv", index=False)
```
