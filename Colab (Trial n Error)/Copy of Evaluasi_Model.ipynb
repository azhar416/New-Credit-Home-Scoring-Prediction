# Import Model

```python
# Load Model 3
LR_Model = pickle.load(open(root + "Model/LR_Model_3.pkl", 'rb'))
RF_Model = pickle.load(open(root + "Model/RF_Model_3.pkl", 'rb'))
KNN_Model = pickle.load(open(root + "Model/KNN_Model_3.pkl", 'rb'))
```

# Evaluasi Model

## Logistik Regression Model

### Model 3

```python
y_pred_lr = LR_Model.predict(X_test.values)
```

```python
print("Accuracy Logit : ", accuracy_score(y_test, y_pred_lr))
print("Precision Logit : ", precision_score(y_test, y_pred_lr, average='macro'))
print("Recall Logit : ",recall_score(y_test, y_pred_lr, average='macro'))
```

```python
score = roc_auc_score(y_test, y_pred_lr)

print(f"ROC AUC LR: {score}")
```

```python
f1_lr = f1_score(y_test, y_pred_lr)
print(f"F1 Score LR : {f1_lr}")
```

```python
# Confusion Matrix Logit Model
print("Cofusion Matrix Logit 2 : \n", confusion_matrix(y_test, y_pred_lr))
```
