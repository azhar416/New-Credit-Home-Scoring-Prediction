# Training

## KNN Model

```python
# KNN
knn_model = KNeighborsClassifier(n_neighbors=7, weights='uniform', leaf_size=50, metric='minkowski')
```

### Hyperparameter Tuning

```python
param_grid = {
    'n_neighbors': [7, 9, 11, 13, 15],
    'weights': ["uniform", "distance"],
    'leaf_size': [30, 40, 50],
    'metric': ["euclidean", "manhattan", "minkowski"],
}
```

```python
knn_Grid = GridSearchCV(estimator = knn_model, param_grid = param_grid, cv = 3, scoring="roc_auc")
```

```python
knn_Grid.fit(X_train.values,y_train.values.ravel())
```

```python
knn_opt = knn_Grid.best_estimator_
```

### Train Model

```python
# knn_opt.fit(X_train.values,y_train.values.ravel())
knn_model.fit(X_train.values,y_train.values.ravel())
```

## Random Forest Model

```python
# Random Forest
rf_model = RandomForestClassifier(random_state=42, n_jobs=-1)
```

### Hyperparameter Tuning

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2,4]
# Minimum number of samples required to split a node
min_samples_split = [2, 5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'bootstrap': bootstrap
}
```

```python
rf_Grid = GridSearchCV(estimator = rf_model, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4, scoring="recall")
```

```python
rf_Grid.fit(X_train, y_train)
```

```python
rf_opt = rf_Grid.best_estimator_
```

### Train Model

```python
rf_opt.fit(X_train,y_train)
```
